```python
import matplotlib.pyplot as plt

data_points = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

dir1 = 'Robust_noise/BN/with_regularization/'
acc1 = []
for data in data_points:
    f = open(dir1 + str(data) + '/details.txt', 'r')
    text = []
    for x in f.readlines():
        text.append(x)
    acc1.append(float(text[-1].split(' ')[-1]))

dir2 = 'Robust_noise/BN/without_regularization/'
acc2 = []
for data in data_points:
    f = open(dir2 + str(data) + '/details.txt', 'r')
    text = []
    for x in f.readlines():
        text.append(x)
    acc2.append(float(text[-1].split(' ')[-1]))

dir3 = 'Robust_noise/non-BN/'
acc3 = []
for data in data_points:
    f = open(dir3 + str(data) + '/details.txt', 'r')
    text = []
    for x in f.readlines():
        text.append(x)
    acc3.append(float(text[-1].split(' ')[-1]))

plt.plot(data_points, acc1, label='BN_reg')
plt.plot(data_points, acc2, label='BN_without_reg')
plt.plot(data_points, acc3, label='No-BN')
plt.xlabel('Data points per class')
plt.ylabel('Best validation accuracy')
plt.legend()
plt.show()
```
